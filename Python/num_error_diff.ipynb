# 微分中的数值计算误差

```python
import matplotlib.pyplot as plt
import numpy as np
```

$$\frac{x}{y}\frac{\mathrm{d} y}{\mathrm{d} x}$$
的误差大于
$$\frac{\mathrm{d} \ln y}{\mathrm{d} \ln x}$$
的误差。

例如下面例子正确答案是$-3$，而用前一种算法的结果有明显的系统误差。

```python
x = np.logspace(1, 3, 100)
y = x**-3
dlnydlnx_num1 = x / y * np.gradient(y, x)
dlnydlnx_num2 = np.gradient(np.log(y), np.log(x))

plt.plot(x, dlnydlnx_num1, label=r'$\frac{x}{y}\frac{\mathrm{d} y}{\mathrm{d} x}$')
plt.plot(x, dlnydlnx_num2, label=r'$\frac{\mathrm{d} \ln y}{\mathrm{d} \ln x}$')
plt.legend(fontsize=14)
```
